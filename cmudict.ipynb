```python
import io
import nltk
from nltk.corpus import cmudict
```

```python
transcr = cmudict.dict()
```

```python
def isVowel(pho):
    if pho[-1] in ["0", "1", "2"]:
        return True
    else:
        return False

def getSyls(arr):
    syls = []
    s = []
    for tok in arr:
        s.append(tok)
        if isVowel(tok):
            syls.append(s)
            s = []
    if len(s) > 0:
        if len(syls) > 0:
            syls[-1] = syls[-1] + s
        else:
            syls = [s]
    return syls
        
```

```python
myFile = open("google-10000-english.txt")
allSyls = []
for line in myFile.readlines():
    word = line.replace("\n","")
    #print(word)
    if word in transcr:
        syls = getSyls(transcr[word][0])
        for syl in syls:
            if syl not in allSyls:
                allSyls.append(syl)
```

```python
for syl in sorted(allSyls):
    print(syl)
```
